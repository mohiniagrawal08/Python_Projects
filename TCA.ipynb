# replacing blank with 0 as tenure is 0 and datatype to string

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ", "0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

# we have converted senior citizen 0/1 into yes/no to make it easier

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
       return "no"

df ["SeniorCitizen"] = df ["SeniorCitizen"].apply(conv)  
```
